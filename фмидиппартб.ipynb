```python
import spacy

nlp = spacy.load("en_core_web_sm")


doc = nlp("I loved the movie.")

spacy_tokens = [token.text for token in doc]
print(spacy_tokens)
```

```python
lowercase_tokens = [token.lower() for token in spacy_tokens]
print(lowercase_tokens)
```

```python
filtered_tokens = [token.text for token in doc if not token.is_stop and not token.is_punct]
print(filtered_tokens)
```

```python
lemmas = [token.lemma_ for token in doc]
print(lemmas)
```

```python
texts = [
    "I loved the movie.",
    "I did not love the movie.",
    "Can you turn the light off?",
    "Apple released a new device in 2025.",
    "Book a flight from Almaty to Astana.",
    "LOL, that was hilarious!"
]

all_tokens = []
for text in texts:
    doc = nlp(text)
    all_tokens.extend([token.text for token in doc if not token.is_stop and not token.is_punct])

total_tokens = len(all_tokens)

unique_tokens = len(set(all_tokens))

from collections import Counter
frequency = Counter(all_tokens)
most_frequent = frequency.most_common(5)

noisy_tokens = [word for word in all_tokens if word in ["light", "off", "was", "that", "a"]]

print("Total Tokens:", total_tokens)
print("Unique Tokens:", unique_tokens)
print("Most Frequent Tokens:", most_frequent)
print("Noisy Tokens:", noisy_tokens)
```

```python
import spacy

try:
    nlp = spacy.load("en_core_web_lg")
    print("Модель 'en_core_web_lg' успешно загружена.")
except OSError:
    print("Модель 'en_core_web_lg' не найдена. Загрузка...")
    !python -m spacy download en_core_web_lg
    print("Загрузка завершена. Перезагрузка модели...")
    nlp = spacy.load("en_core_web_lg")

words = ["happy", "sad", "angry", "actor", "film", "director", "plane", "train", "airport"]

neighbors = {}

for word in words:

    word_vec = nlp(word).vector

    similar_words = nlp.vocab.vectors.most_similar(word_vec.reshape(1, -1), n=6)

    similar_words_list = [nlp.vocab.strings[i] for i in similar_words[0][0] if nlp.vocab.strings[i] != word]

    # Добавляем в словарь
    neighbors[word] = similar_words_list

for word, neighbor in neighbors.items():
    print(f"Neighbors of '{word}': {neighbor}")
```
